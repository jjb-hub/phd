```python
import numpy as np
import matplotlib.pyplot as plt
import math

def direct_distance(h):
    return math.sqrt((1/4) + h*h)  

def reflected_distance(h, num_bounces=1):
    way_out = (math.sqrt((h*h/9) + 1/4)) 
    way_in = (math.sqrt((4*h*h/9) + 1))
    intermediary_bounce = 2 * way_in
    return way_out + num_bounces * intermediary_bounce + way_in

def sphere_surface(r):
    return 4*math.pi*r*r

def sound_volume(d):
    return 1 - 1 / sphere_surface(d)

heigths = np.linspace(0, 10, 400) # in meters
reflections = list(range(10)) # in meters


plt.figure(figsize=(8, 6))
plt.plot(list[range(100)], [sound_volume(d) for d in range(100)], label=f"sound volume per distane")
plt.xlabel("distance")
plt.ylabel("volume")
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)  # Add x-axis line
plt.axvline(0, color='black',linewidth=0.5)  # Add y-axis line
plt.legend()
plt.title("Sound volume per reflection")
plt.show()
```

```python

direct_volume = [sound_volume(direct_distance(h)) for h in heigths]

plt.figure(figsize=(8, 6))
for height in range(10):
    volume = [sound_volume(direct_distance(height))]+[sound_volume(reflected_distance(height, num_reflected_waves)) for num_reflected_waves in reflections[:-1]]
    plt.plot(reflections, volume, label=f"{height}m")
plt.xlabel("nth reflection")
plt.ylabel("volume")
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)  # Add x-axis line
plt.axvline(0, color='black',linewidth=0.5)  # Add y-axis line
plt.legend()
plt.title("Sound volume per reflection")
plt.show()
```

```python

direct_volume = [sound_volume(direct_distance(h)) for h in heigths]


plt.figure(figsize=(8, 6))
for height in range(10):
    volumes = [sound_volume(direct_distance(height))]
    for num in range(99):
        volumes.append(volumes[-1] + sound_volume(reflected_distance(height, num)))
    plt.plot(list(range(100)), volumes, label=f"{height}m")
    
plt.xlabel("num reflections")
plt.ylabel("cumulative volume")
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)  # Add x-axis line
plt.axvline(0, color='black',linewidth=0.5)  # Add y-axis line
plt.legend()
plt.title("Sound volume per reflection")
plt.show()
```

```python

total_volume = [sound_volume(direct_distance(h)) for h in heigths]

plt.figure(figsize=(8, 6))

plt.plot(heigths, total_volume, label="direct sound volume")
for num_reflected_waves in range(10):
    additional_volume = [sound_volume(reflected_distance(h, num_reflected_waves)) for h in heigths]
    total_volume = np.sum([total_volume, additional_volume], axis = 0)
    plt.plot(heigths, total_volume, label=f"{num_reflected_waves + 1} reflections")
plt.xlabel("height")
plt.ylabel("volume")
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)  # Add x-axis line
plt.axvline(0, color='black',linewidth=0.5)  # Add y-axis line
plt.title("Cumulative sound volume")
plt.legend()
plt.show()
```

```python

for height in range(20):
    plt.plot(reflections + [len(reflections) + 1], [sound_volume(direct_distance(height))] + [sound_volume(reflected_distance(height, num_reflected_waves)) for num_reflected_waves in reflections], label=f"height {height}")
plt.xlabel("number of reflections")
plt.ylabel("volume")
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)  # Add x-axis line
plt.axvline(0, color='black',linewidth=0.5)  # Add y-axis line
plt.legend()
plt.show()

```
